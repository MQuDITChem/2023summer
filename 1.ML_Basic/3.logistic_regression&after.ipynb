```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Logistic (regression) classfication

```python
# set seed
np.random.seed(123)

X1 = 5 * np.random.rand(5, 1)
X2 = 5 * np.random.rand(5, 1) + 5
X = np.concatenate((X1, X2), axis=0)
y = np.concatenate((np.zeros(5), np.ones(5)), axis=0) 
```

```python
df = pd.DataFrame({'study_time': X.reshape(-1), 'pass': y})
df
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.scatter(df["study_time"][:5], df["pass"][:5], color='k', marker='^')
ax.scatter(df["study_time"][5:], df["pass"][5:], color='r', marker='^')
ax.scatter(14, 1, color='b', marker='x')
ax.scatter(14, 14 * 1.3 / 20, color='g', marker='o')
ax.scatter(5.5, 5.5 * 2.0 / 20, color='y', marker='o')
ax.scatter(5.5, 5.5 * 1.3 / 20, color='g', marker='o')
ax.scatter(5.5, 1, color='b', marker='x')

ax.set_xlabel('Study time', fontsize=14, fontweight='bold')
ax.set_ylabel('Score', fontsize=14, fontweight='bold')
ax.set_yticks([-0.5, 0, 0.5, 1, 1.5])
ax.set_yticklabels([-0.5, '(Fail) 0', 0.5, '(pass) 1', 1.5])
ax.plot([0, 20], [0.0, 2], color='y', linestyle='--', label='y = 0.1x + 0.0')
ax.plot([0, 20], [0.0, 1.3], color='g', linestyle='--', label='y = 0.065x + 0.0')
ax.plot([0, 20], [0.5, 0.5], color='grey', linestyle='--', linewidth=0.5)

ax.set_xlim(0, 15)
ax.set_ylim(-0.5, 1.5)
plt.legend()
plt.show()
```
